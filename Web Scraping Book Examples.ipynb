```python
import requests
import bs4
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
base_url.format(20)
```

```python
req = requests.get(base_url.format(1))
```

```python
soup = bs4.BeautifulSoup(req.text,'lxml')
```

```python
soup.select(".product_pod")
```

```python
products = soup.select(".product_pod")
```

```python
first_book = products[0]
```

```python
three_stars_rated_books = []

for page_no in range(1,51):
    
    scrape_url = base_url.format(page_no)  
    
    req = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(req.text,'lxml')
    
    books = soup.select(".product_pod")
    
    for book in books:
        
        if "star-rating Three" in str(book):
            
            book_title = book.select('a')[1]['title']
            
            three_stars_rated_books.append(book_title)
```

```python
three_stars_rated_books
```

```python
imp_url = "https://books.toscrape.com/catalogue/category/books_1/page-{}.html"
```

```python
five_stars_rated_books = []

for page_no in range(1,51):
    
    scrape_url = imp_url.format(page_no)  
    
    req = requests.get(scrape_url)
    
    soup = bs4.BeautifulSoup(req.text,'lxml')
    
    books = soup.select(".product_pod")
    
    for book in books:
        
        if len(book.select(".star-rating.Five")) != 0:
            
            book_title = book.select('a')[1]['title']
            
            five_stars_rated_books.append(book_title)
```
